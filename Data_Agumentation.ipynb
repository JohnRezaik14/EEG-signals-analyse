```python
import pandas as pd
import numpy as np
```

```python
uploaded_file_name = list(uploaded.keys())[0]
```

```python
data = pd.read_csv(uploaded_file_name)

# Display the first few rows of the data
print(data)
data_types = data.dtypes
print(data_types)
```

```python


# Additive noise parameters
noise_mean = 0
noise_std = 0.1  # Adjust this value based on the desired noise level
print(data)
num_data = data.iloc[0:, 1:-1]
print(num_data)
# Generate random noise
noise = np.random.normal(noise_mean, noise_std, num_data.shape)

# Add noise to EEG data
eeg_with_noise = num_data + noise
first_column = data.iloc[:, 0:1].copy()
last_column = data.iloc[:, -1:].copy()
eeg_with_noise=pd.concat([first_column,eeg_with_noise,last_column],axis=1)
print(eeg_with_noise)
```
