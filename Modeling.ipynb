### CNN

```python
st_act="tanh"
# st_act="softsign"
nd_act='softmax'
opt=Adam
learning_rate=2e-5
loss="sparse_categorical_crossentropy"
epochs=150
batch_size=4
validation_split=0.1

modelCNN = Sequential()
modelCNN.add(Conv1D(filters=24, kernel_size=3, activation=st_act, input_shape=(304, 1)))
modelCNN.add(LeakyReLU(alpha=0.1))
modelCNN.add(Conv1D(filters=16, kernel_size=3, activation=st_act))
modelCNN.add(LeakyReLU(alpha=0.1))
modelCNN.add(Conv1D(filters=8, kernel_size=3, activation=st_act))
modelCNN.add(LeakyReLU(alpha=0.1))
modelCNN.add(Conv1D(filters=4, kernel_size=3, activation=st_act))
modelCNN.add(LeakyReLU(alpha=0.1))
modelCNN.add(MaxPooling1D(pool_size=2))
modelCNN.add(Flatten())
modelCNN.add(Dense(units=128, activation=st_act))
modelCNN.add(LeakyReLU(alpha=0.1))
modelCNN.add(Dense(units=64, activation=st_act))
modelCNN.add(LeakyReLU(alpha=0.1))
modelCNN.add(Dropout(0.5))
modelCNN.add(Dense(units=3, activation=nd_act))

modelCNN.compile(optimizer=opt(learning_rate=learning_rate),
              loss=loss,
              metrics=['accuracy'])

history=modelCNN.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.1)
```

```python
# Evaluate the model
test_loss, test_accuracy = modelCNN.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
# 3/3 [==============================] - 0s 8ms/step - loss: 0.5699 - accuracy: 0.8485
# Test Loss: 0.5699
# Test Accuracy: 0.8485
# Test Loss: 0.0306
# Test Accuracy: 1.0000 epoch 80
# Test Loss: 0.0841
# Test Accuracy: 1.0000 epoch 100
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.2, 1.5])
plt.legend(loc='lower right')
```

```python

results.append({
 "st_act":st_act,
"nd_act":nd_act,
"optimizer":opt,
"learning_rate": learning_rate,
"loss":loss,
"epochs":epochs,
"batch_size":batch_size,
"validation_split":validation_split,
     "test loss":test_loss,
    "Test Accuracy":test_accuracy
    })
```
